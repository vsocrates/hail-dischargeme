```python
from datasets import load_dataset

# If the dataset is gated/private, make sure you have run huggingface-cli login
configs = ['1. Discharge Summary Sections (Targets)', '2. Discharge Summaries', '3. Radiology Reports', '4. ED Stays', '5. ED Triages', '6. ED Diagnoses']
datasets = []
for config in configs:
    dataset = load_dataset("justin13601/discharge-me", config)
    datasets.append(dataset)
```

```python
target_train = pd.DataFrame(datasets[0]['train'])
target_valid = pd.DataFrame(datasets[0]['valid'])
target_test = pd.DataFrame(datasets[0]['test'])

discharge_summ_train = pd.DataFrame(datasets[1]['train'])
discharge_summ_valid = pd.DataFrame(datasets[1]['valid'])
discharge_summ_test = pd.DataFrame(datasets[1]['test'])

radiology_train = pd.DataFrame(datasets[2]['train'])
radiology_valid = pd.DataFrame(datasets[2]['valid'])
radiology_test = pd.DataFrame(datasets[2]['test'])

edstay_train = pd.DataFrame(datasets[3]['train'])
edstay_valid = pd.DataFrame(datasets[3]['valid'])
edstay_test = pd.DataFrame(datasets[3]['test'])

edtriage_train = pd.DataFrame(datasets[4]['train'])
edtriage_valid = pd.DataFrame(datasets[4]['valid'])
edtriage_test = pd.DataFrame(datasets[4]['test'])

eddiags_train = pd.DataFrame(datasets[5]['train'])
eddiags_valid = pd.DataFrame(datasets[5]['valid'])
eddiags_test = pd.DataFrame(datasets[5]['test'])
```

```python
target_train['split'] = "train"
target_valid['split'] = "valid"
target_test['split'] = "test"
discharge_summ_train['split'] = "train"
discharge_summ_valid['split'] = "valid"
discharge_summ_test['split'] = "test"
radiology_train['split'] = "train"
radiology_valid['split'] = "valid"
radiology_test['split'] = "test"
edstay_train['split'] = "train"
edstay_valid['split'] = "valid"
edstay_test['split'] = "test"
edtriage_train['split'] = "train"
edtriage_valid['split'] = "valid"
edtriage_test['split'] = "test"
eddiags_train['split'] = "train"
eddiags_valid['split'] = "valid"
eddiags_test['split'] = "test"
```

```python
target = pd.concat([target_train, target_valid, target_test])
discharge_summ = pd.concat([discharge_summ_train, discharge_summ_valid, discharge_summ_test])
radiology = pd.concat([radiology_train, radiology_valid, radiology_test])
edstay = pd.concat([edstay_train, edstay_valid, edstay_test])
edtriage = pd.concat([edtriage_train, edtriage_valid, edtriage_test])
eddiags = pd.concat([eddiags_train, eddiags_valid, eddiags_test])
```

```python
target = target.reset_index()
discharge_summ = discharge_summ.reset_index()
radiology = radiology.reset_index()
edstay = edstay.reset_index()
edtriage = edtriage.reset_index()
eddiags = eddiags.reset_index()
```
