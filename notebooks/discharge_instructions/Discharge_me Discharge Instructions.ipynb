```python
def zero_shot_azure(prompt, note):
    client = AzureOpenAI(
      azure_endpoint = "https://decile-openai-llm.openai.azure.com/", 
      api_key="bfdc4ba059264591a99da53f0fbb71ca",  
      api_version="2024-02-15-preview"
    )


    message_text = [{"role":"system","content":prompt+"'''"+note+"'''"}]
    completion = client.chat.completions.create(
      model="decile-gpt-35-turbo-16k", # model = "deployment_name"
      messages = message_text,
      temperature=0.7,
      max_tokens=800,
      top_p=0.95,
      frequency_penalty=0,
      presence_penalty=0,
      stop=None
    )
    return completion.choices[0].message.content
```

```python
def multi_shot_azure(prompt, note):
    client = AzureOpenAI(
      azure_endpoint = "https://decile-openai-llm.openai.azure.com/", 
      api_key="bfdc4ba059264591a99da53f0fbb71ca",  
      api_version="2024-02-15-preview"
    )

    message_text = [
        {"role":"system","content":"You are a physician tasked to write discharge instructions for a patient based on their corresponding hospital course. First, you will be presented with three pairs of hospital courses and their discharge instructions. Use these examples to inform the discharge instructions you will write. You will then be given a hospital course for you to write the discharge instructions for."},
        {"role":"system","content":"Hospital Course: "+ df_discharge_target.iloc[1]["brief_hospital_course"]+ "And the appropriate discharge instructions accompanying the hospital course: "+df_discharge_target.iloc[1]["discharge_instructions"]},
        {"role":"system","content":"Hospital Course: "+ df_discharge_target.iloc[2]["brief_hospital_course"]+ "And the appropriate discharge instructions accompanying the hospital course: "+df_discharge_target.iloc[2]["discharge_instructions"]},
        {"role":"system","content":"Hospital Course: "+ df_discharge_target.iloc[6]["brief_hospital_course"]+ "And the appropriate discharge instructions accompanying the hospital course: "+df_discharge_target.iloc[6]["discharge_instructions"]},
        {"role":"system","content":prompt+"'''"+note+"'''"}]

    completion = client.chat.completions.create(
      model="decile-gpt-35-turbo-16k", # model = "deployment_name"
      messages = message_text,
      temperature=0.7,
      max_tokens=800,
      top_p=0.95,
      frequency_penalty=0,
      presence_penalty=0,
      stop=None
    )
    return completion.choices[0].message.content
```

```python
target_filtered = df_discharge_target[(df_discharge_target['discharge_instructions_word_count'] > 10) & (df_discharge_target['brief_hospital_course_word_count'] > 10)]
target_filtered.reset_index(inplace = True, drop = True)
target_filtered.head(10)
```

```python
df_discharge_target.iloc[0]["discharge_instructions"]
```

```python
output_zero_1 = zero_shot_azure(current_prompt, target_filtered.iloc[0]["brief_hospital_course"])
output_multi_1 = multi_shot_azure(current_prompt, target_filtered.iloc[0]["brief_hospital_course"])
```

```python
output_zero_1
```

```python
output_multi_1
```

```python
dc_results = pd.DataFrame(columns = ["ZeroShot Output", "MultiShot Output", "Original"])
dc_results
```

```python
for i in target_filtered.index[:10]:
    try:
        output_zero = zero_shot_azure(current_prompt, target_filtered.iloc[i]["brief_hospital_course"])
        output_multi = multi_shot_azure(current_prompt, target_filtered.iloc[i]["brief_hospital_course"])
    except Exception as e:
        print(f"An error occurred: {e}")
    temp = {
        "ZeroShot Output": output_zero,
        "MultiShot Output": output_multi,
        "Original": target_filtered.loc[i]["discharge_instructions"]
    }
    df_temp = pd.DataFrame(temp, index = [0])
    dc_results = pd.concat([dc_results, df_temp], ignore_index=True)
```

```python
from rouge_score import rouge_scorer
```

```python
def rouge_scores(output, gold_standard):
    scorer = rouge_scorer.RougeScorer(['rouge1',  "rouge2", 'rougeL'], use_stemmer=True)
    scores = scorer.score(output, gold_standard)
    return scores
```

```python
rouge_scores(output_multi_1, target_filtered.iloc[0]["discharge_instructions"])
```

```python
for i in dc_results.index:
    print("Zero Shot Scores:")
    print(rouge_scores(dc_results.loc[i]["ZeroShot Output"], dc_results.loc[i]["Original"]))
    print("Multi Shot Scores:")
    print(rouge_scores(dc_results.loc[i]["MultiShot Output"], dc_results.loc[i]["Original"]))
```
